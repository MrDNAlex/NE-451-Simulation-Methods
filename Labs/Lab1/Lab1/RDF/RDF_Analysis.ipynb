# Load Data from the file


```python
import pandas
import matplotlib.pyplot as plt

def read_RDF_File (path : str):
    df = pandas.read_csv(path, skiprows=4, delim_whitespace=True, header=None)
    columns = [
    "Bin", "r",
    "g_Li-Li", "coord_Li-Li",
    "g_Li-Mn", "coord_Li-Mn",
    "g_Li-O", "coord_Li-O"
    ]
    
    df.columns = columns
    
    return df

RDF_Data_20 = read_RDF_File("Data/RDF_20.data")
RDF_Data_100 = read_RDF_File("Data/RDF_100.data")
RDF_Data_100_2 = read_RDF_File("Data/RDF_100_2.data")
RDF_Data_100_3 = read_RDF_File("Data/RDF_100_3.data")

```

```python
def plot_RDF (points : int, df : pandas.DataFrame, title : str):
    # Plot g(r) for Li-Li, Li-Mn, Li-O
    plt.figure(figsize=(20, 16))

    plt.plot(df["r"][-points:], df["g_Li-Li"][-points:], label="Li-Li")
    plt.plot(df["r"][-points:], df["g_Li-Mn"][-points:], label="Li-Mn")
    plt.plot(df["r"][-points:], df["g_Li-O"][-points:], label="Li-O")

    plt.xlabel("r (Ã…)")
    plt.ylabel("g(r)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_RDF(20, RDF_Data_20, "Radial Distribution Function (RDF) (T = 670K)")
```

```python
plot_RDF(100, RDF_Data_100, "Radial Distribution Function (RDF) (T = 300K)")
plot_RDF(100, RDF_Data_100_2, "Radial Distribution Function (RDF) (T = 300K)")
plot_RDF(100, RDF_Data_100_3, "Radial Distribution Function (RDF) (T = 300K)")
```
