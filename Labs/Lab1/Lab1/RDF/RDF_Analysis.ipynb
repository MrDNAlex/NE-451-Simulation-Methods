# Load Data from the file


```python
import pandas
import matplotlib.pyplot as plt

RDF_Data_20 = pandas.read_csv("RDF.data", skiprows=4, delim_whitespace=True, header=None)

RDF_Data_20.columns = [
    "Bin", "r",
    "g_Li-Li", "coord_Li-Li",
    "g_Li-Mn", "coord_Li-Mn",
    "g_Li-O", "coord_Li-O"
]

print(RDF_Data_20)
```

```python
import matplotlib.pyplot as plt

# Plot g(r) for Li-Li, Li-Mn, Li-O
plt.figure(figsize=(20, 16))

plt.plot(RDF_Data_20["r"][-20:], RDF_Data_20["g_Li-Li"][-20:], label="Li-Li")
plt.plot(RDF_Data_20["r"][-20:], RDF_Data_20["g_Li-Mn"][-20:], label="Li-Mn")
plt.plot(RDF_Data_20["r"][-20:], RDF_Data_20["g_Li-O"][-20:], label="Li-O")

plt.xlabel("r (Å)")
plt.ylabel("g(r)")
plt.title("Radial Distribution Function (RDF)")
plt.legend()
plt.grid(True)
plt.show()
```
